```python
#used to  modify the state of a function or add additional functionality to an function

import time

def get_time(func):  #get_time is the name of your decorators
    def wrapper(args):
        start_time=time.time()
        res=func(args)
        end_time=time.time()
        print('Time taken:'+str((end_time-start_time)*1000)+'ms' )
        
        return res
    return wrapper


@get_time

def squares(a_list):
    sq=[]
    for ele in a_list:
        sq.append(ele**2)
    return sq

@get_time
def cubes(a_list):
    cu=[]
    for ele in _list:
        cu.append(ele**3)
    return cu
squares([1,2,3,4,5,6])
#cubes([1,2,3,4])
```
